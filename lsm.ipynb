```python
from quantkit.pricing import *
```

```python
am_call = Option(strike=100, maturity=1.0, option_type=OptionType.PUT, style=OptionStyle.AMERICAN)
eu_put = Option(strike=100, maturity=1.0, option_type=OptionType.PUT, style=OptionStyle.EUROPEAN)
```

```python
# Define Market State
market = MarketData(spot=100.0, rate=0.05, volatility=0.7)

# Configure and Initialize Pricer
config = LSMCConfig(n_paths=100000, n_steps=100, degree=3)
pricer = LSMCEngine(config)

# Execute Pricing
price_am = pricer.price(am_call, market)
price_eu = pricer.price(eu_put, market)
```

```python
plot_exercise_boundaries(
    instrument=am_call,
    lsmc_res=price_am, 
    benchmark_res=None # Optional
)
```
